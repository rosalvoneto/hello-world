```
import numpy as np
import matplotlib.pyplot as plt
```

```
v1 = np.array([1, 3])
v2 = np.array([3, 4])

v3 = v2 - v1

print(v3)
```

```
fig, ax = plt.subplots(figsize = (10, 10)) # Tamanho da imagem 10 x 10

# Min e Max X e Y
plt.ylim(0,7)
plt.xlim(0,7)

ax.arrow(0, 0, v1[0], v1[1], ec='r')
ax.arrow(0, 0, v2[0], v2[1], ec='b')

ax.arrow(v1[0], v1[1], v3[0], v3[1], ec='y')

plt.show()
```
